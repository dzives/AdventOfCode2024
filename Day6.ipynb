```python
from aocd import get_data

# https://adventofcode.com/2024/day/6
sample_data = """....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""

def parse(sample=False):
    if not sample:
        x = get_data(day=6,year=2024)
    else:
        x = sample_data
    y = [list(i) for i in x.splitlines()]
    # PARSE THE INPUT
    return y
```

```python
from collections import Counter

def get_starting_position(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == '^':
                return [i,j]
    return None

def part1(sample=False):
    matrix = parse(sample)
    current_pos = get_starting_position(matrix)
    dir_i = 0
    dirs = [[-1,0],[0,1],[1,0],[0,-1]] # up, right,down,left
    visited = []
    while True:
        if current_pos not in visited:
            visited.append(current_pos[:])
        next_move = [current_pos[0] +  dirs[dir_i][0], current_pos[1] + dirs[dir_i][1]]
        if (next_move[0] < 0 or next_move[0] >= len(matrix) or next_move[1] < 0 or next_move[1] >= len(matrix[0])):
            break
        if matrix[next_move[0]][next_move[1]] == "#":
            dir_i+=1
            dir_i = dir_i % len(dirs)
        else:
            current_pos = next_move
    return len(visited)





def part2(sample=False):
    matrix = parse(sample)
    current_pos = get_starting_position(matrix)
    dir_i = 0
    dirs = [[-1,0],[0,1],[1,0],[0,-1]] # up, right,down,left
    visited = []
    out = 0
    while True:
        if current_pos not in visited:
            visited.append(current_pos)
        next_move = [current_pos[0] +  dirs[dir_i][0], current_pos[1] + dirs[dir_i][1]]
        if (next_move[0] < 0 or next_move[0] >= len(matrix) or next_move[1] < 0 or next_move[1] >= len(matrix[0])):
            break
        if matrix[next_move[0]][next_move[1]] == "#":
            dir_i+=1
            dir_i = dir_i % len(dirs)
        else:
            current_pos = next_move
   
    for i in visited:
        current_pos = get_starting_position(matrix)
        vis = []
        dir_i = 0
        while True:
            vis.append(current_pos)
            next_move = [current_pos[0] +  dirs[dir_i][0], current_pos[1] + dirs[dir_i][1]]
            if (next_move[0] < 0 or next_move[0] >= len(matrix) or next_move[1] < 0 or next_move[1] >= len(matrix[0])):
                break
            if matrix[next_move[0]][next_move[1]] == "#" or (next_move[0] == i[0] and next_move[1] == i[1]) :
                dir_i+=1
                dir_i = dir_i % len(dirs)
            else:
                current_pos = next_move
            if len(vis) > len(matrix) * len(matrix[0]):
                out+=1
                break

    return out    
```

```python
from time import time
from termcolor import colored
import os

if __name__ == '__main__':
    p1c = 41
    p2c = 6
    print(f'sample: ')
    start = time()
    p1r = part1(True)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(True)
        end = time()
        p2t = end-start
    if p1c == p1r:
        print(colored('Part 1 sample is correct.', 'green'))
    else:
        print(colored('Part 1 sample is incorrect.', 'red'))
    if p2c is None:
        print(colored('Part 2 sample is not defined.', 'yellow'))
    elif p2c == p2r:
        print(colored('Part 2 sample is correct.', 'green'))
    else:
        print(colored('Part 2 sample is incorrect.', 'red'))
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
    print(f'input: ')
    start = time()
    p1r = part1(False)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(False)
        end = time()
        p2t = end-start
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
```
