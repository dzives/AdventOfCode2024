```python
from aocd import get_data

sample_data = """MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX"""

def parse(sample=False):
    if not sample:
        x = get_data(day=4,year=2024)
    else:
        x = sample_data
    rows = x.splitlines()
    matrix = [list(row) for row in rows]
    cols = list(zip(*rows)) 
    cols = [''.join(col) for col in cols]
    # \ diagonals
    diag1 = []
    for i in range(len(matrix)):
        temp = []
        j = 0
        while True:
            try:
                temp.append(matrix[i+j][j])
                j+=1
            except IndexError:
                break
        diag1.append(''.join(temp))
    for i in range(len(matrix[0])):
        temp = []
        j = 0
        while True:
            try:
                temp.append(matrix[j][i+j+1])
                j+=1
            except IndexError:
                break
        diag1.append(''.join(temp))

    # / diagonals
    matrix = [list(row[::-1]) for row in rows]
    diag2 = []
    for i in range(len(matrix)):
        temp = []
        j = 0
        while True:
            try:
                temp.append(matrix[i+j][j])
                j+=1
            except IndexError:
                break
        diag2.append(''.join(temp))
    for i in range(len(matrix[0])):
        temp = []
        j = 0
        while True:
            try:
                temp.append(matrix[j][i+j+1])
                j+=1
            except IndexError:
                break
        diag2.append(''.join(temp))
    matrix = [list(row.replace("X",".")) for row in rows]
    # PARSE THE INPUT
    return [*rows,*cols,*diag1,*diag2],matrix
```

```python
from regex import findall

def part1(sample=False):
    joined,_ = parse(sample)
    
    target = "XMAS"
    out = 0
    for i in joined:
        out+= len(findall(target,i))
        out+= len(findall(target[::-1],i))

    return out

def part2(sample=False):
    _,matrix = parse(sample)
    target = ["MAS","SAM"]
    out = 0
    for i in range(1,len(matrix[0])):
        for j in range(1,len(matrix[i])):
            if matrix[i][j] == 'A':
                try:
                    d1 = matrix[i-1][j-1] + matrix[i][j] + matrix[i+1][j+1]
                    d2 = matrix[i-1][j+1] + matrix[i][j] + matrix[i+1][j-1]
                    if d1 in target and d2 in target:
                        out+=1
                except IndexError:
                    continue
    return out
```

```python
from time import time
from termcolor import colored
import os

if __name__ == '__main__':
    p1c = 18
    p2c = 9
    print(f'sample: ')
    start = time()
    p1r = part1(True)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(True)
        end = time()
        p2t = end-start
    if p1c == p1r:
        print(colored('Part 1 sample is correct.', 'green'))
    else:
        print(colored('Part 1 sample is incorrect.', 'red'))
    if p2c is None:
        print(colored('Part 2 sample is not defined.', 'yellow'))
    elif p2c == p2r:
        print(colored('Part 2 sample is correct.', 'green'))
    else:
        print(colored('Part 2 sample is incorrect.', 'red'))
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
    print(f'input: ')
    start = time()
    p1r = part1(False)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(False)
        end = time()
        p2t = end-start
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
```
