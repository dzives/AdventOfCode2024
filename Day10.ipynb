```python
from aocd import get_data

# https://adventofcode.com/2024/day/10
sample_data = """89010123
78121874
87430965
96549874
45678903
32019012
01329801
10456732"""

def parse(sample=False):
    if not sample:
        x = get_data(day=10,year=2024)
    else:
        x = sample_data
    return [list(map(int, list(i))) for i in x.splitlines()]
```

```python
height = 0
width = 0
out = []

def traverse(matrix,i,j,expected):
    if i < 0 or i >= height or j < 0 or j >= height:
        return 0
    if matrix[i][j] != expected:
        return 0
    if matrix[i][j] == 9 and expected == 9:
        # print(f"\t[{i},{j}] = {matrix[i][j]}")
        if [i,j] not in out:
            out.append([i,j])
        return 1
    return traverse(matrix,i+1,j,expected+1) + traverse(matrix,i,j+1,expected+1) + traverse(matrix,i-1,j,expected+1) +traverse(matrix,i,j-1,expected+1)

def part1(sample=False):
    global height
    global width
    global out
    in_data = parse(sample)
    height = len(in_data)
    width = len(in_data[0])
    res = 0
    for i in range(height):
        for j in range(width):
            if in_data[i][j] == 0:
                traverse(in_data,i,j,0)
                res+=len(out)
                out.clear()
    return res


def part2(sample=False):
    global height
    global width
    in_data = parse(sample)
    height = len(in_data)
    width = len(in_data[0])
    res = 0
    for i in range(height):
        for j in range(width):
            if in_data[i][j] == 0:
                res+= traverse(in_data,i,j,0)
    return res
```

```python
from time import time
from termcolor import colored
import os

if __name__ == '__main__':
    p1c = 36
    p2c = 81
    print(f'sample: ')
    start = time()
    p1r = part1(True)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(True)
        end = time()
        p2t = end-start
    if p1c == p1r:
        print(colored('Part 1 sample is correct.', 'green'))
    else:
        print(colored('Part 1 sample is incorrect.', 'red'))
    if p2c is None:
        print(colored('Part 2 sample is not defined.', 'yellow'))
    elif p2c == p2r:
        print(colored('Part 2 sample is correct.', 'green'))
    else:
        print(colored('Part 2 sample is incorrect.', 'red'))
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
    print(f'input: ')
    start = time()
    p1r = part1(False)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(False)
        end = time()
        p2t = end-start
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
```
