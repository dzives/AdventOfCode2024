```python
from aocd import get_data

# https://adventofcode.com/2024/day/9
sample_data = """2333133121414131402"""

def parse(sample=False):
    if not sample:
        x = get_data(day=9,year=2024)
    else:
        x = sample_data
    if len(x) % 2 ==1:
        x = x + '0'
    y = list(zip(list(map(int, x.strip()[::2])),list(map(int, x.strip()[1::2]))))
    return y
```

```python
def part1(sample=False):
    in_data = parse(sample)
    out = 0
    count = 0
    mem = []
    for data,space in in_data:
        mem = mem + [count for _ in range(data)] + ['.' for _ in range(space)]
        count+=1
    empty = mem.count(".")
    index = len(mem) - 1
    while True:
        if index == -1 or empty == 0:
            break
        if index != ".":
            mem[index],mem[mem.index(".")] = mem[mem.index(".")],mem[index]
            empty += -1
        index += -1
    index = 0
    while mem[index] != '.':
        out+=index*mem[index]
        index += 1
    return out

def combine_spaces(mem):
    index = 0
    while index < len(mem) - 1:
        if mem[index][0] == '.' and mem[index + 1][0] == '.':
            mem[index] = ('.', mem[index][1] + mem[index + 1][1])
            del mem[index + 1]
        else:
            index += 1
    return mem


def print_mem(mem):
    a = ["."*i[1] if i[0] == "." else str(i[0])*i[1] for i in mem]
    return ''.join(a)

def part2(sample=False):
    in_data = parse(sample)
    out = 0
    count = 0
    mem = []
    for data, space in in_data:
        mem = mem + [(count, data)] + ([(".", space)] if space != 0 else [])
        count += 1
    empty_index = lambda: list(((i, v) for i, v in enumerate(mem) if v[0] == '.'))

    for index in range(len(mem)-1, -1, -1):
        if mem[index][0] == ".":
            continue
            
        size = mem[index][1]
        moved = False
        
        for i in range(len(mem)):
            if i >= index:
                break
                
            if mem[i][0] == "." and mem[i][1] >= size:
                new_size = mem[i][1] - size
                mem[i] = mem[index]
                mem.insert(i + 1, (".", new_size))
                mem[index+1] = (".",mem[index+1][1])
                
                index -= abs(len(mem) - len(combine_spaces(mem)))
                mem = combine_spaces(mem)
                break
    
    index = 0
    for i,type in enumerate([item[0] for i in mem for item in [i] * i[1]]):
        if type == '.':
            continue
        out += i*int(type)
    return out
```

```python
from time import time
from termcolor import colored
import os

if __name__ == '__main__':
    p1c = 1928
    p2c = 2858
    print(f'sample: ')
    start = time()
    p1r = part1(True)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(True)
        end = time()
        p2t = end-start
    if p1c == p1r:
        print(colored('Part 1 sample is correct.', 'green'))
    else:
        print(colored('Part 1 sample is incorrect.', 'red'))
    if p2c is None:
        print(colored('Part 2 sample is not defined.', 'yellow'))
    elif p2c == p2r:
        print(colored('Part 2 sample is correct.', 'green'))
    else:
        print(colored('Part 2 sample is incorrect.', 'red'))
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
    print(f'input: ')
    start = time()
    p1r = part1(False)
    end = time()
    p1t = end-start
    if p2c:
        start = time()
        p2r = part2(False)
        end = time()
        p2t = end-start
    print(f'Part 1: {p1r}|Time: {p1t:.3f}s')
    if p2c:
        print(f'Part 2: {p2r}|Time: {p2t:.3f}s')
```
